## **Data Preparation and Loading**

```python
import pandas as pd
sales_data=pd.read_csv('/content/sales_data.csv')

```

```python
print(sales_data)
```

## **Data Cleaning: Handling Missing Values**

```python
# Check for missing values
print("Missing values in the dataset:")
print(sales_data.isnull().sum())
```

```python
# Handling missing values by filling with mean, median, or a specific value
sales_data['turnover'].fillna(sales_data['turnover'].median(), inplace=True)
sales_data.head(20)
```

```python
# Drop rows with missing values in other columns if needed
sales_data.dropna(inplace=True)
sales_data
```

```python
# Verify that missing values have been handled
print("\nAfter handling missing values:")
print(sales_data.isnull().sum())
sales_data
```

## **Data Cleaning: Handling Outliers**

```python
from scipy import stats
```

```python
# Calculate the z-scores for 'turnover' to identify outliers
z_scores = stats.zscore(sales_data['turnover'])
z_scores
```

```python
# Define a threshold for outliers (e.g., z-score > 3 or < -3)
outlier_threshold = 3
outlier_threshold
```

```python
# Identify outlier indices
outlier_indices = abs(z_scores) > outlier_threshold
outlier_indices.head()
```

```python
# Remove outliers
sales_data = sales_data[~outlier_indices]
sales_data.head()
```

```python
# Verify that outliers have been handled
print("\nAfter handling outliers:")
print(sales_data.shape)
```

in the above code:
* We calculate the z-scores for the 'Sales' column in the dataset.
* We set a threshold of 3 for the absolute z-score to identify outliers.
*
We identify and remove the rows with 'Sales' z-scores exceeding the threshold, effectively removing the outliers from the dataset.

## **Exploratory Analysis: Summary Statistics**

```python
# Display summary statistics of the numerical columns
summary_stats = sales_data.describe()
print(summary_stats)
```

```python
# Display unique values and their counts for categorical columns
unique_values = sales_data['Product'].value_counts()
print("\nUnique values for 'Product':")
print(unique_values)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of 'turnover' to visualize the distribution
plt.figure(figsize=(10, 6))
sns.histplot(sales_data['turnover'], kde=True, color='skyblue')
plt.title('Distribution of turnover')
plt.xlabel('turnover')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot of 'Sales' to visualize outliers
plt.figure(figsize=(20, 6))
sns.boxplot(x=sales_data['Product'], y=sales_data['turnover'], palette='pastel')
plt.title('Sales by Product')
plt.xlabel('Product')
plt.ylabel('turnover')
plt.xticks(rotation=45)
plt.show()
```

In the above code:
* We first display summary statistics such as count, mean, min, max, and
quartiles for numerical columns like 'Sales'.
* We then visualize the distribution of 'Sales' using a histogram with a kernel density estimate.
* We also use a box plot to visualize the spread and potential outliers in 'Sales' for each product.

```python
# Most sold products
most_sold_products = sales_data['Product'].value_counts().idxmax()
print("Most sold product:" ,{most_sold_products})
```

```python
# Total sales and average sales
total_sales = sales_data['turnover'].sum()
average_sales = sales_data['turnover'].mean()
print(f"Total sales: ${total_sales:.2f}")
print(f"Average sales per transaction: ${average_sales:.2f}")
```

```python
# Trends over time (e.g., monthly sales trends)
monthly_sales_trends = sales_data.resample('M').sum()['turnover']
print("\nMonthly Sales Trends:")
print(monthly_sales_trends)
```

In this example:

* We identify the most sold product based on the highest frequency in the 'Product' column.
* We determine the highest sales month by aggregating sales on a monthly basis and finding the month with the maximum sales.
* We calculate the total sales and average sales per transaction.
* We display the monthly sales trends to observe any patterns or trends over time.
